```python
import joblib
import os
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import subprocess
```

```python
def create_study(char) -> list:
    study = joblib.load(f"{char}_study.pkl")
    best_trial = study.best_trial
    param_keys = list(best_trial.params.keys())
    param_values = {key: list() for key in param_keys}
    scores = list()

    for trial in study.trials:
        for key in param_keys:
            param_values[key].append(trial.params[key])
        scores.append(trial.value)

    # show scatter
    for key in param_keys:
        plt.figure()
        plt.scatter(param_values[key], scores, s=8)
        plt.xlabel(key)
        plt.ylabel("acc score")
        plt.show()
```

```python
create_study("vit1d")
```

```python
create_study("convbbt")
```
