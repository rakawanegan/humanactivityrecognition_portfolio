「歩く」は1/2秒の周期でピークがあり、ｘ軸は他のｙｚ軸より常に小さい値をとっている。  
「走る」は1/4秒の周期でピークが存在し、ｙ軸について負の値が「歩く」より顕著である。  
「下り階段」ではｙ軸の加速度が小さく、1/2秒の周期がある。  
「上り階段」ではｙｚ軸に3/4秒の周期があり、これは足の進みが他より遅いことを表している。  

```python
TIME_PERIODS = 80
SAMPLING_RATE = 20
TIME = TIME_PERIODS / SAMPLING_RATE
GRAVITY = 9.81
```

```python
def standing(TIME, GRAVITY=9.81):
    base = np.linspace(0, TIME, 80)
    x = 0 * base
    y = np.tile(GRAVITY, 80)
    z = 0 * base
    return np.array([x, y, z]).T

def sitting(TIME, GRAVITY=9.81):
    base = np.linspace(0, TIME, 80)
    x = np.tile(GRAVITY*0.1, 80)
    y = np.tile(GRAVITY*0.1, 80)
    z = np.tile(GRAVITY*0.8, 80)
    return np.array([x, y, z]).T

def walking(TIME, GRAVITY=9.81):
    base = np.linspace(0, TIME, 80)
    x = np.sin(2 * np.pi * (1/0.5) * base)
    y = np.sin(2 * np.pi * (1/0.5) * base)
    z = np.sin(2 * np.pi * (1/0.5) * base)
    return np.array([x, y, z]).T

def jogging(TIME, GRAVITY=9.81):
    base = np.linspace(0, TIME, 80)
    x = np.sin(2 * np.pi * (1/0.25) * base)
    y = np.sin(2 * np.pi * (1/0.25) * base)
    z = np.sin(2 * np.pi * (1/0.25) * base)
    return np.array([x, y, z]).T

def ascending(TIME, GRAVITY=9.81):
    base = np.linspace(0, TIME, 80)
    x = np.sin(2 * np.pi * (1/0.75) * base)
    y = np.sin(2 * np.pi * (1/0.75) * base)
    z = np.sin(2 * np.pi * (1/0.75) * base)
    return np.array([x, y, z]).T

def descending(TIME, GRAVITY=9.81):
    base = np.linspace(0, TIME, 80)
    x = np.sin(2 * np.pi * (1/0.5) * base)
    y = np.sin(2 * np.pi * (1/0.5) * base)
    z = np.sin(2 * np.pi * (1/0.5) * base)
    return np.array([x, y, z]).T

def rondom(TIME, SEED=314):
    np.random.seed(SEED)
    x = np.random.rand(80)
    y = np.random.rand(80)
    z = np.random.rand(80)
    return np.array([x, y, z]).T
```

```python
plot_activity("standing", standing(TIME))
```
